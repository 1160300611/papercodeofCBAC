```python
num_gene = len(genes)
num_dis = len(diss)
num_symp = len(symps)
num_go = len(gos)

dim  = num_gene + num_dis + num_go + num_symp
```

```python
type_mask = np.zeros(dim, dtype=int)
type_mask[num_gene:num_gene+num_dis] = 1
type_mask[num_gene+num_dis:num_gene+num_dis+num_go] = 2
type_mask[num_gene+num_dis+num_go:] = 3

adjM = np.zeros((dim, dim), dtype=int)
```

```python
for _, row in ppi.iterrows():
    gid1 = gene_dict[row[0]]
    gid2 = gene_dict[row[1]]
    adjM[gid1, gid2] = 1
    adjM[gid2, gid1] = 1
```

```python
for _, row in dis_gene.iterrows():
    did = num_gene + dis_dict[row[0]]
    gid = gene_dict[row[1]]
    adjM[did, gid] = 1
    adjM[gid, did] = 1
```

```python
print(num_gene, num_dis, num_go, num_symp)
```

```python
for _, row in gene_go.iterrows():
    gid = gene_dict[row[0]]
    goid = num_gene + num_dis + go_dict[row[1]]
    adjM[gid, goid] = 1
    adjM[goid, gid] = 1
```

```python
for _, row in dis_symp.iterrows():
    did = num_gene + dis_dict[row[0]]
    sid = num_gene + num_dis + num_go + symp_dict[row[1]]
    adjM[did, sid] = 1
    adjM[sid, did] = 1
```

```python
# 0-1
gene_dis_list = {i: adjM[i, num_gene:num_gene+num_dis].nonzero()[0] for i in range(num_gene)}
# 1-0
dis_gene_list = {i: adjM[i, :num_gene].nonzero()[0] for i in range(num_dis)}
# 0-2
gene_go_list = {i: adjM[i, num_gene+num_dis:num_gene+num_dis+num_go].nonzero()[0] for i in range(num_gene)}
# 2-0
go_gene_list = {i: adjM[i, :num_gene].nonzero()[0] for i in range(num_go)}
# 0-0
gene_gene_list = {i: adjM[i, :num_gene].nonzero()[0] for i in range(num_gene)}
# 1-3
dis_symp_list = {i: adjM[i, num_gene+num_dis+num_go:].nonzero()[0] for i in range(num_dis)}
# 3-1
symp_dis_list = {i: adjM[i, num_gene:num_gene+num_dis].nonzero()[0] for i in range(num_symp)}
```

```python
# 0-1-0
g_d_g = []
for d, g_list in dis_gene_list.items():
    g_d_g.extend([(g1, d, g2) for g1 in g_list for g2 in g_list])
g_d_g = np.asarray(g_d_g)
g_d_g[:, 1] += num_gene
sorted_index = sorted(list(range(len(g_d_g))), key=lambda i : g_d_g[i, [0, 2, 1]].tolist())
g_d_g = g_d_g[sorted_index]

# 0-2-0
g_o_g = []
for o, g_list in go_gene_list.items():
    g_o_g.extend([(g1, o, g2) for g1 in g_list for g2 in g_list])
g_o_g = np.asarray(g_o_g)
g_o_g[:, 1] +=num_gene + num_dis
sorted_index = sorted(list(range(len(g_o_g))), key=lambda i : g_o_g[i, [0, 2, 1]].tolist())
g_o_g = g_o_g[sorted_index]

# 1-0-1
d_g_d = []
for g, d_list in gene_dis_list.items():
    d_g_d.extend([(d1, g, d2) for d1 in d_list for d2 in d_list])
d_g_d = np.asarray(d_g_d)
d_g_d += num_gene
d_g_d[:, 1] -= num_gene
sorted_index = sorted(list(range(len(d_g_d))), key=lambda i : d_g_d[i, [0, 2, 1]].tolist())
d_g_d = d_g_d[sorted_index]

```

```python
# 1-0-0-1
d_g_g_d = []
for g1, g_list in gene_gene_list.items():
    for g2 in g_list:
        d_g_g_d.extend([(d1, g1, g2, d2) for d1 in gene_dis_list[g1] for d2 in gene_dis_list[g2]])
d_g_g_d = np.asarray(d_g_g_d)
d_g_g_d[:, [0, 3]] += num_gene
sorted_index = sorted(list(range(len(d_g_g_d))), key=lambda i : d_g_g_d[i, [0, 3, 1, 2]].tolist())
d_g_g_d = d_g_g_d[sorted_index]

# 1-3-1
d_s_d = []
for s, d_list in symp_dis_list.items():
    d_s_d.extend([(d1, s, d2) for d1 in d_list for d2 in d_list])
d_s_d = np.asarray(d_s_d)
d_s_d += num_gene
d_s_d[:, 1] += num_dis + num_go
sorted_index = sorted(list(range(len(d_s_d))), key=lambda i : d_s_d[i, [0, 2, 1]].tolist())
d_s_d = d_s_d[sorted_index]
```

```python
# 0-0
g_g = []
for g1, g_list in gene_gene_list.items():
    g_g.extend([(g1, g2) for g2 in g_list])
g_g = np.asarray(g_g)
sorted_index = sorted(list(range(len(g_g))), key=lambda i : g_g[i, [0, 1]].tolist())
g_g = g_g[sorted_index]
```

```python
expected_metapaths = [
    [(0, 1, 0), (0, 2, 0), (0, 0)],
    [(1, 0, 1), (1, 0, 0, 1), (1, 3, 1)]
]

for i in range(len(expected_metapaths)):
    pathlib.Path(save_prefix + '{}'.format(i)).mkdir(parents=True, exist_ok=True)

metapath_indices_mapping = {(0, 1, 0): g_d_g,
                           (0, 2, 0): g_o_g,
                           (0, 0): g_g,
                           (1, 0, 1): d_g_d,
                           (1, 0, 0, 1): d_g_g_d,
                           (1, 3, 1): d_s_d}

target_idx_lists = [np.arange(num_gene), np.arange(num_dis)]
offset_list = [0, num_gene]
for i, metapaths in enumerate(expected_metapaths):
    for metapath in metapaths:
        edge_metapath_idx_array = metapath_indices_mapping[metapath]
        
        with open(save_prefix + '{}/'.format(i) + '-'.join(map(str, metapath)) + '_idx.pickle', 'wb') as out_file:
            target_metapaths_mapping = {}
            left = 0
            right = 0
            for target_idx in target_idx_lists[i]:
                while right < len(edge_metapath_idx_array) and edge_metapath_idx_array[right, 0] == target_idx + offset_list[i]:
                    right += 1
                target_metapaths_mapping[target_idx] = edge_metapath_idx_array[left:right, ::-1]
                left = right
            pickle.dump(target_metapaths_mapping, out_file)
            
        with open(save_prefix + '{}/'.format(i) + '-'.join(map(str, metapath)) + '.adjlist', 'w') as out_file:
            left = 0
            right = 0
            for target_idx in target_idx_lists[i]:
                while right < len(edge_metapath_idx_array) and edge_metapath_idx_array[right, 0] == target_idx + offset_list[i]:
                    right += 1
                neighbors = edge_metapath_idx_array[left:right, -1] - offset_list[i]
                neighbors = list(map(str, neighbors))
                if len(neighbors) > 0:
                    out_file.write('{} '.format(target_idx) + ' '.join(neighbors) + '\n')
                else:
                    out_file.write('{}\n'.format(target_idx))
                left = right
                
scipy.sparse.save_npz(save_prefix + 'adjM.npz', scipy.sparse.csr_matrix(adjM))
np.save(save_prefix + 'node_types.npy', type_mask)
```

```python
num_gene = 21584
num_dis = 15030
gene_dis_edges = np.load(save_prefix + 'gene_dis_edges.npy')
train_val_test_idx = np.load(save_prefix + 'train_val_test_idx.npz')
train_idx = train_val_test_idx['train_idx']
val_idx = train_val_test_idx['val_idx']
test_idx = train_val_test_idx['test_idx']
```

```python
print('----------Generate val and test netgative sample----------')
neg_candidates = []
counter = 0
for i in range(num_gene):
    for j in range(num_dis):
        if counter < len(gene_dis_edges):
            if i == gene_dis_edges[counter, 0] and j == gene_dis_edges[counter, 1]:
                counter += 1
            else:
                neg_candidates.append([i, j])
        else:
            neg_candidates.append([i, j])
neg_candidates = np.array(neg_candidates)
print('----------Finished----------')

idx = np.random.choice(len(neg_candidates), len(val_idx) + len(test_idx), replace=False)
val_neg_candidates = neg_candidates[sorted(idx[:len(val_idx)])]
test_neg_candidates = neg_candidates[sorted(idx[len(val_idx):])]
train_neg_candidates = np.vstack((neg_candidates, gene_dis_edges[val_idx], gene_dis_edges[test_idx]))
```

```python
print(f'val negative samples: {val_neg_candidates.shape[0]:}')
print(f'test negative samples: {test_neg_candidates.shape[0]:}')
print(f'train negative samples: {train_neg_candidates.shape[0]:}')
```

```python
np.savez(save_prefix + 'train_val_test_neg_gene_dis.npz',
         train_neg_gene_dis=train_neg_candidates,
         val_neg_gene_dis=val_neg_candidates,
         test_neg_gene_dis=test_neg_candidates)
np.savez(save_prefix + 'train_val_test_pos_gene_dis.npz',
         train_pos_gene_dis=gene_dis_edges[train_idx],
         val_pos_gene_dis=gene_dis_edges[val_idx],
         test_pos_gene_dis=gene_dis_edges[test_idx])
```
